```python
import numpy as np

A = np.array([[3,4],
              [7,8]])

B = np.array([[5,3],
              [2,1]])
```

```python
import numpy as np

A = np.array([[3,4],[7,8]])

A_inv = np.linalg.inv(A)
identity = A @ A_inv

print("A inverse:\n", A_inv)
print("A * A_inv:\n", identity)
```

```python
import numpy as np

A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

AB = A @ B
BA = B @ A

print("AB =\n", AB)
print("BA =\n", BA)
print("Are AB and BA equal?:", np.array_equal(AB, BA))
```

```python
import numpy as np

A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

lhs = (A @ B).T
rhs = B.T @ A.T

print("(AB)^T =\n", lhs)
print("B^T A^T =\n", rhs)
print("Are they equal?:", np.array_equal(lhs, rhs))
```
