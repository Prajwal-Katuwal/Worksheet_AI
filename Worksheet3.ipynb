Solve the linear systems graphically.
2x+3y=3
4x+6y=6
3x-2y=1


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the x range
x_vals = np.linspace(-10, 10, 400)

# Define the equations solved for y
y1 = (3 - 2*x_vals) / 3
y2 = (6 - 4*x_vals) / 6
y3 = (3*x_vals - 1) / 2

# Plot the lines
plt.plot(x_vals, y1, label='2x + 3y = 3')
plt.plot(x_vals, y2, label='4x + 6y = 6')
plt.plot(x_vals, y3, label='3x - 2y = 1')

# Solving equations 1 and 3 as 1 and 2 are dependent (they're the same line)
# 2x + 3y = 3
# 3x - 2y = 1

from sympy import symbols, Eq, solve

x, y = symbols('x y')
eq1 = Eq(2*x + 3*y, 3)
eq2 = Eq(3*x - 2*y, 1)
sol = solve((eq1, eq2), (x, y))

# Plot the point of intersection
if sol:
    xi, yi = float(sol[x]), float(sol[y])
    plt.plot(xi, yi, 'ro')  # Red dot at the intersection
    plt.text(xi + 0.2, yi, f'({xi:.2f}, {yi:.2f})', color='red')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphical Solution of Linear System')
plt.show()
```

3.	Create a 4x4 matrix containing random numbers in the range of 1 to 20. Have the user prompted for a number, check that the imputed value is a number and then check if the number is in the matrix. If the number is in the matrix, return the number of times that number is in the matrix.

```python
# Create 4x4 matrix with random numbers from 1 to 20
matrix = np.random.randint(1, 21, (4, 4))
print("Matrix:\n", matrix)

# Prompt user for a number
user_input = input("\nEnter a number to search in the matrix (1-20): ")

# Validate input
if user_input.isdigit():
    number = int(user_input)
    if 1 <= number <= 20:
        count = np.count_nonzero(matrix == number)
        if count > 0:
            print(f"\nThe number {number} is found {count} time(s) in the matrix.")
        else:
            print(f"\nThe number {number} is not found in the matrix.")
    else:
        print("Number out of range. Please enter between 1 and 20.")
else:
    print("Invalid input. Please enter a numeric value.")
```

1.	Create a matrix  of order 5  5 with random entries. Calculate the rank of the matrix  to check the number of linearly independent columns or rows and to check if the matrix is invertible. What are the other ways to check if the matrix is invertible?



```python
import numpy as np

# Create a 5x5 matrix A with random integers
A = np.random.randint(1, 10, (5, 5))
print("Matrix A:\n", A)

# Calculate rank
rank = np.linalg.matrix_rank(A)
print("\nRank of A:", rank)

# Check if invertible: must be full rank (rank = 5)
if rank == 5:
    print("Matrix A is invertible.")
else:
    print("Matrix A is not invertible.")

try:
    det = np.linalg.det(A)
    print("Determinant of A:", det)
except np.linalg.LinAlgError:
    print("Determinant could not be computed.")
```

2.	Generate a matrix  of order 5  5 with random entries and generate a vector (column)  of length 5 with random entries. Calculate Matrix-vector product  = . Now, we have a linear system, i.e. the matrix  and a vector . Solve this Linear System.

```python
# Generate random 5x5 matrix A and 5x1 vector x
A = np.random.randint(1, 10, (5, 5))
x = np.random.randint(1, 10, (5, 1))

# Matrix-vector product
b = A @ x

print("Matrix A:\n", A)
print("\nVector x:\n", x)
print("\nVector b (A @ x):\n", b)

# Solve A*x = b
x_solution = np.linalg.solve(A, b)
print("\nSolution (solving A*x = b):\n", x_solution)
```

3.	Solve the following system.
8x+3y2z=9
4x+7y+5z==15
3x+4y12z=35


```python
from sympy import symbols, Eq, solve, init_printing

# Pretty print for equations
init_printing()

# Define variables
x, y, z = symbols('x y z')

# Define equations
eq1 = Eq(8*x + 3*y - 2*z, 9)
eq2 = Eq(-4*x + 7*y + 5*z, 15)
eq3 = Eq(3*x + 4*y - 12*z, 35)

# Solve system
solution = solve((eq1, eq2, eq3), (x, y, z))
print("Solution to the system:\n", solution)
```

Task 4:  
1.	Consider the following 10 points in X-Y plane: (3, 7.6), (2,1), (1, 4.8), (0, 5), (1, 2.8), (2, 0.6), (3, 4), (4, 6.2), (5, 6), (7, 6.4). Plot these points in X-Y plane. Connect the points to guess the polynomial. Once you guess your polynomial, formulate a Linear System and then solve it to find the coefficient of the polynomial. Finally, check if the polynomial fits along with the above given points.


```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the given points
x_vals = np.array([-3, -2, -1, 0, 1, 2, 3, 4, 5, 7])
y_vals = np.array([-7.6, 1, 4.8, 5, 2.8, -0.6, -4, -6.2, -6, 6.4])

# Step 2: Plot the original points
plt.scatter(x_vals, y_vals, color='red', label='Given Points')
plt.plot(x_vals, y_vals, color='gray', linestyle='dotted', label='Point Connections')
plt.title("Given 10 Points in X-Y Plane")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.legend()

# Step 3: Guess polynomial of degree 9 (since 10 points)
# Create Vandermonde matrix for degree 9 polynomial
V = np.vander(x_vals, 10)

# Step 4: Solve linear system V * a = y_vals
coeffs = np.linalg.solve(V, y_vals)

# Step 5: Plot fitted polynomial
x_fit = np.linspace(min(x_vals), max(x_vals), 500)
y_fit = np.polyval(coeffs, x_fit)

plt.plot(x_fit, y_fit, label='Fitted Polynomial (deg 9)', color='blue')
plt.legend()
plt.show()

# Step 6: Print the polynomial coefficients
print("Polynomial Coefficients (from highest degree to constant):\n", coeffs)
```
