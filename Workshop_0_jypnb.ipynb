**Task 1**


```python
time_data = [
    (3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0),
    (4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
    (5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
    (3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
    (4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0)
]
low = []
moderate = []
high = []
for item in time_data:
    value = item[0]
    if value < 3:
        low.append(item)
    elif value < 5:
        moderate.append(item)
    else:
        high.append(item)
print("Low:", low)
print("Moderate:", moderate)
print("High:", high)
```

**Task 2**


```python
l_count = 0
for i in low:
  l_count += len(i)
print(f"The total value in low is {l_count}")

m_count = 0
for i in moderate:
  m_count += len(i)
print(f"The total value in moderate is {m_count}")

h_count = 0
for i in high:
  h_count += len(i)
print(f"The total value in high is {h_count}")
```

**Task 3**


```python
study_minutes = []

for i in time_data:
    minutes = tuple(x * 60 for x in i)
    study_minutes.append(minutes)
print(study_minutes)
```

**Task 4**

```python
study_hours = []
entertainment_hours = []
sleep_hours = []


for i in time_data:
  study_hours.append(i[0])
  entertainment_hours.append(i[1])
  sleep_hours.append(i[2])

study_average = sum(study_hours)/len(study_hours)
entertainment_average = sum(entertainment_hours)/len(entertainment_hours)
sleep_average = round(sum(sleep_hours)/len(sleep_hours),1)

print(f'The average study time is {study_average}')
print(f'The average entertainment time is {entertainment_average}')
print(f'The average sleep time is {sleep_average}')
```

```python
import matplotlib.pyplot as plt
time_data = [
    (3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0),
    (4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
    (5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
    (3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
    (4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0)
]
study_hours = [i[0] for i in time_data]
sleep_hours = [i[2] for i in time_data]
plt.figure(figsize=(8,5))
plt.scatter(study_hours, sleep_hours, color='purple')
plt.xlabel("Study Hours")
plt.ylabel("Sleep Hours")
plt.title("Study vs Sleep Pattern")
plt.show()
```
